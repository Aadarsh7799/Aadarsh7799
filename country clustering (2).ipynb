```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig=data.plot.scatter(x='Country',y='Language')
```

```python
new_data=data.iloc[:,1:3]
```

```python
new_data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(new_data)
```

```python
clusters=kmeans.fit_predict(new_data)
```

```python
print(clusters)
```

```python
sns.scatterplot(x=new_data.iloc[:, 0], y=new_data.iloc[:, 1], hue=clusters)
```
