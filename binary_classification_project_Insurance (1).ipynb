```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/insurance2 (1).csv')
```

```python
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='bmi',y='charges',data=data,hue='insuranceclaim')
```

```python
sns.relplot(x='age',y='charges',data=data,hue='insuranceclaim')
```

```python
x=data[['age','sex','bmi','children','smoker','region','charges']]
y=data['insuranceclaim']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
prediction=logistic.predict(x_test)
```

```python
print(prediction)
```

```python
logistic.score(x,y)#R2 score
```

```python
sns.relplot(x=x_test['bmi'], y=y_test, data=x_test, hue=prediction)
```
