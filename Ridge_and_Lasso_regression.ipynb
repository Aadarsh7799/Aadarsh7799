```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/insurance.csv')
```

```python
data.head()
```

Data Cleaning

```python
data['sex']=data['sex'].apply({'female':1,'male':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southwest':1,'southeast':2,'northwest':3,'northeast':4}.get)
```

```python
data.head()
```

Dividing data into dependent and independent



```python
x=data[['age','sex','bmi','children','smoker','region']]#independent
y=data['expenses'] #dependent
```

Spliting the data into train and test set


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
predictions=regression.predict(x_test)# future predictions of MLM
```

```python
regression.score(x,y)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=1.0) #(y-yhat)^2+1
ridge.fit(x_train,y_train)
```

```python
ridge_predictions=ridge.predict(x_test)
```

```python
print(ridge_predictions)
```

```python
ridge.score(x,y)
```

```python
lasso=Lasso(alpha=0.5)
lasso.fit(x_train,y_train)
```

```python
lasso_predictions=lasso.predict(x_test)
```
